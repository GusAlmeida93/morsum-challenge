```python
import pandas as pd
import numpy as np
import math
```

```python
def pack_size(row):
    if math.isnan(row['pack_size']):
        return float(row['pack_size_text'].split(' ')[0])
    else:
        return row['pack_size']
    
def item_weight(row):
    if math.isnan(row['item_weight']):
        if math.isnan(row['item_density']):
            return float(row['pack_size_text'].split(' ')[3])
        else:
            return row['item_volume'] * row['item_density']    
    else:
        return row['item_weight']
    
def pack_size_text(row):
    if type(row['pack_size_text']) == float:
        return f"{int(row['pack_size'])} packs of {row['item_weight']} kg"
    else:
        return row['pack_size_text']
```

```python
df = pd.read_csv('ingredient_quantities.csv')
```

```python
df
```

```python
df['pack_size'] = df.apply(pack_size, axis=1)
```

```python
df
```

```python
df['item_weight'] = df.apply(item_weight, axis=1)
```

```python
df
```

```python
df['total_weight'] = df['pack_size'] * df['item_weight']
```

```python
df
```

```python
df['pack_size_text'] = df.apply(pack_size_text, axis=1)
```
